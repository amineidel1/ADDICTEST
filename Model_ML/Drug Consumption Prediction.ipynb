# Analyse des données

La base de données contient les enregistrements de 1885 personnes. Pour chaque personne, 12 attributs sont connus : 

**Mesures de la personnalité :** NEO-FFI-R (névrosisme, extraversion, ouverture à l'expérience, amabilité et conscience), BIS-11 (impulsivité) et ImpSS (recherche de sensations) </br>
**Informations Personelles :** niveau d'éducation, l'âge, le sexe, le pays de résidence et l'origine ethnique.  

De plus, les participants ont été interrogés sur leur consommation de 18 drogues légales et illégales. </br>
**Drogues Légals :** alcool, benzodiazépine, nicotine, caféine, chocolat. </br>
**Drogues Illégals :** amphétamines, nitrite d'amyle, cannabis, cocaïne, crack, ecstasy, héroïne, kétamine, legalh, LSD, méthadone, Mushroom, l'abus de substances volatiles.</br>
**Drogue fictif :** Semeron - qui a été introduit pour identifier les sur-réclamants. </br>
Pour chaque drogue, ils doivent sélectionner l'une des réponses : n'a jamais utilisé, l'a utilisé il y a plus d'une décennie ou au cours de la dernière décennie , année, mois, semaine ou jour.

## Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('max_columns', None) # Afficher tous les colonnes
sns.set(style="white", color_codes=True) # utliliser le style de sns
```

```python
data = pd.read_csv('data/Drug_Consumption.csv')
data = data.drop('ID', axis=1) # Supprimer la colonne "ID" parcequ'il n'as pas de valeur pour l'analyse
data.head()
```

```python
# Chercher les valeurs manquantes
data.isna().sum().sum()
```

La base de données ne contient aucune valeurs manquante

```python
print("Nombre d'enregistrement original : ", data.shape[0])
```

```python
print("Nombre d'enregistrements non voulu : ", data[data['Semer']!='CL0'].shape[0])
```

```python
# Supprimer la colonne et les entrés des personne ayant dit qu'il ont consomer la drogue fictif "Semer"
data = data.drop(data[data['Semer'] != 'CL0'].index)
data = data.drop('Semer', axis=1)
data = data.drop('Choc', axis=1)
data = data.drop('Caff', axis=1)
data = data.drop('Benzos', axis=1)
data.reset_index()
print("Nombre d'enregistrement sans la drogue fictif : ", data.shape[0])
```

## Data Encoding

```python
data['Gender'] = data['Gender'].map({'M':1,'F':0})
```

```python
ordinal_features = ['Age','Education','Alcohol','Amyl','Amphet','Cannabis','Coke','Crack',
                    'Ecstasy','Heroin','Ketamine','Legalh','LSD','Meth','Mushrooms','Nicotine','VSA']
ordinal_orderings = [
    ['18-24', '25-34', '35-44', '45-54', '55-64', '65+'],
    ['Left school before 16 years','Left school at 16 years','Left school at 17 years','Left school at 18 years','Some college or university, no certificate or degree',
     'Professional certificate/ diploma','University degree','Masters degree','Doctorate degree'],
    ['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],
    ['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],
    ['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],
    ['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],
    ['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6'],['CL0','CL1','CL2','CL3','CL4','CL5','CL6']
]
nominal_features = ['Country','Ethnicity']
```

```python
# Odinal encoding
def ordinal_encoder(df, columns, oredrings):
    df = df.copy()
    for column, ordering in zip(columns, oredrings):
        df[column] = df[column].apply(lambda x : ordering.index(x))
    return df
# nominal/categorical encoding
def cat_converter(df, columns):
    df = df.copy()
    for column in columns:
        df[column] = df[column].astype('category').cat.codes
    return df
```

```python
data = ordinal_encoder(data, ordinal_features, ordinal_orderings)
data = cat_converter(data, nominal_features)
data.head()
```

```python
data.dtypes
```

```python
data = data.astype('float64')
data.head()
```

```python
drugs = ['Alcohol','Amyl','Amphet','Cannabis','Coke','Crack','Ecstasy','Heroin','Ketamine','Legalh','LSD','Meth','Mushrooms','Nicotine','VSA' ]
drugs_mean = []
for drug in drugs:
    drugs_mean.append(data[drug].mean())
drug_use = pd.DataFrame(np.c_[drugs,drugs_mean], columns=['drug','average_use'])
drug_use['average_use'] = drug_use['average_use'].astype('float64')
drug_use = drug_use.sort_values(['average_use'], ascending=False)
```

Comme prévue on constate que l'Alcool, la Nicotine et le Canabis sont les drogues les plus concommées.

**Conclusion**

### Correlation entre les scores psychologiques

```python
legal_corr = legal_df.corr()
legal_corr = legal_corr.drop(legal_drugs, axis=0)
legal_corr = legal_corr.drop(legal_drugs, axis=1)
plt.figure(figsize=(16,8))
sns.heatmap(legal_corr, annot=True, vmin=-1, vmax=1)
plt.title('Psycological Correlations', size=16)
plt.xlabel('Personal Info', size=12)
plt.ylabel('Personal Info', size=12)
plt.show()
```

```python
both_df = legal_df.copy()
both_df['Nic_User'] = both_df['Nicotine'].apply(lambda x: 0.5 if x not in [0,1] else 0) 
both_df['Alco_User'] = both_df['Alcohol'].apply(lambda x: 0.5 if x not in [0,1] else 0)
both_df['Both'] = both_df[['Nic_User', 'Alco_User']].iloc[:].sum(axis=1)
both_df.head()
# 0.0 = non nicotine or benzo consumer
# 0.5 = consumes either drug
# 1.0 = consumes both drugs
plt.figure(figsize=(14,8))
sns.scatterplot(data=both_df, x='Cscore', y='Nscore', hue='Both')
plt.title('Relationship between Neuroticism and Conscientiousness')
plt.xlabel('NEO Five-Factor Inventory Conscientiousness score')
plt.ylabel('NEO Five-Factor Inventory Neuroticism score')
plt.show()
```

**Conclusion**

Fait intéressant, ont a remarqué une forte corrélation négative entre le Nscore et la Conscience (Cscore). Pour mieux comprendre comment cette relation était liée à la consommation de drogue, j'ai examiné la consommation de nicotine et de l'alcohol. Sans surprise, ceux qui consommaient à la fois de la nicotine et de l'alcohol avaient tendance à avoir des Nscores plus élevés et des Cscores inférieurs, tandis que ceux qui n'utilisaient aucune drogue affichaient des Cscores élevés et des Nscores faibles.

# Prédiction de consomation des drogues

## Feature Engineering

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics 

from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
# Supprimer les features avec faible corrélation
low_corr = ['Age', 'Gender', 'Education','AScore']
for column in low_corr:
    data = data.drop(column, axis=1)
    personal_info.remove(column)
```

```python
for drug in legal_drugs+illegal_drugs:
    data[drug] = data[drug].apply(lambda x: 1 if x not in [0,1] else 0)
data.head()
```

## Model Predecting

```python
def predict_drug_usage(model ,drug):
    X_train, X_test, y_train, y_test = train_test_split(data[personal_info], data[drug], test_size=0.2, random_state=1)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    print("Accuracy Score de","[",drug,"] :",model.score(X_test,y_test))
    print("F1 Score de","[",drug,"] :",metrics.f1_score(y_test,y_predict))
```

### Naive Bayes

```python
model = GaussianNB()
for drug in legal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
for drug in illegal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
```

### Random Forest Classifier

```python
model = RandomForestClassifier()
for drug in legal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
for drug in illegal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
```

### SVM Classifier

```python
model = svm.SVC()
for drug in legal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
for drug in illegal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
```

### Logistic Regression

```python
model = LogisticRegression()
for drug in legal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
for drug in illegal_drugs:
    predict_drug_usage(model, drug)
    print('---------------------')
```

```python
tuned_parameters = [{"solver" : ['liblinear'],"multi_class" : ['ovr'],
        "class_weight" : [None, 'balanced'],"max_iter" : [100,200,500,800, 1600]},
        {"solver" : ['newton-cg'],"multi_class" : ['ovr','multinomial'],
        "class_weight" : [None, 'balanced'],"max_iter" : [100,200,500,800, 1600]}]
model_cv = GridSearchCV(model, tuned_parameters, cv=10,scoring='f1')
for drug in legal_drugs:
    predict_drug_usage(model_cv, drug)
    print('---------------------')
for drug in illegal_drugs:
    predict_drug_usage(model_cv, drug)
    print('---------------------')
```
